```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston') 
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis,color="magenta", label='Boston')
# Invert Y Axis so January is on top.
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
#Add a title and axes labels
ax.set_title("PyberFare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create Plot, Change Color to Red and Legend for Chicago
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
# Title Axes and Labels
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('Pyber Fare by Month')
# Add a limit for the x-axis data
plt.xlim(0, 50)
# Add the Legend
plt.legend
# Invert the y-axis so Jan at top
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use Python List comprehension technique to replace for and while loops
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# S is size parameter 
fig, ax = plt.subplots()
# Change the color of the markers to sky blue.
ax.scatter(y_axis, x_axis, s=y_axis, label='Boston')
# Change the size of the markers to 5 times each data point,make color skyblue,Make the color 20% transparent
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c='skyblue', alpha=0.8, edgecolors='black')
#Add a title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
# Set limit for X axis.
ax.set_xlim(0, 50)
# Invert the y-axis so Jan at top
ax.invert_yaxis()
```

```python
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# S is size parameter 
fig, ax = plt.subplots()
colors = ['slategrey', 'blue', 'indigo', 'grey', 'teal', 'aqua', 'yellow', 'cyan', 'teal', 'hotpink', 'violet', 'plum']  
fig, ax = plt.subplots()
ax.pie(y_axis, x_axis, s=y_axis)
# explode two highest percentage months
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
        explode=explode_values)
plt.show()
```
